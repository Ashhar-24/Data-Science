# Logical Operator


`and` : returns `True` when both the conditions are true

`or` : when either of the condition is true

`not`: takes negation of the given condition


```python
print((10 > 2) or (10 < 12))
print(not True)
```

- `Note` for operation: `and`

When all the operands are true, output will be the `last operand` as the and operator works from left to right

eg:

 5 `and` 4 `and` 10

output: 10

# Bitwise Operator

```python
print(10 | 4)
```

# - Loop and Iteration


# while loop


```python
# printing even no.s

i = 1
while (i <= 20):
    if (i % 2 == 0):
        print(i)
    i += 1
```

- `end=" "`: For printing everything in one line


```python
i = 1

while (i <= 20):
    if (i % 2 == 0):
        # print(i, end=" ") #prints with space
        print(i, end=", ")
    i += 1
```

# for loop


```python
# range (a,b) => [a,b), step size: 1 (default)

# range (a,b,c) => [a,b), step size: c
```

```python
for i in range(1, 21):
    if (i % 2 == 0):
        print(i, end=" ")
```

```python
for i in range(2, 21, 2):
    print(i, end=" ")
```

```python
# Printing multiplication table for a given no.

num = 4

for i in range(1, 11):
    print(num, "x", i, "=", num*i)
```

# Break and Continue


- `break` : Exit all the iterations
- `continue` : Skip the current iteration and move to next iteration


```python
for i in range(1, 11):
    if (i == 5):
        break
    print(i, end=" ")
```

```python
for i in range(1, 11):
    if (i == 5):
        continue # will ignore all the statements below and move to next iteration
    print(i, end=" ")
```
